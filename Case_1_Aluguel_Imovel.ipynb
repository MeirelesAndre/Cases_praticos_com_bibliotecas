```python
#IMPORTANTE A BLIBLIOTECA PANDAS
import pandas as pd
```

```python
#PUXANDO A PLANILHA HOUSES_RENT
aluguel_casa = pd.read_excel("houses_rent.xlsx")
```

```python
#PUXANDO DADOS DE CIMA PARA BAIXO
aluguel_casa.head()
```

```python
#PUXANDO DADOS DE BAIXO PARA CIMA
aluguel_casa.tail()
```

```python
#RENOMEANDO ALGUNS TITULOS DE COLUNAS
aluguel_casa = aluguel_casa.rename(columns={'parking spaces':'parking_spaces','rent amount':'rent_amount','property tax':'property_tax','fire insurance':'fire_insurance',})
```

```python
#PUXANDO DADOS COM TITULO RENOMEADO
aluguel_casa.head()
```

```python
#PUXANDO OS TIPOS DE CADA COLUNA
aluguel_casa.dtypes
```

```python
#ALTERANDO ALGUMAS COLUNAS PARA TIPO FLOAT
aluguel_casa = aluguel_casa.astype({"rent_amount":"float","property_tax":"float","fire_insurance":"float"})
aluguel_casa.dtypes
```

```python
#VERIFICANDO AS INFORMAÇÕES COMPLETAS
aluguel_casa.info()
```

```python
#DESCRIÇÃO DE ALGUMAS COLUNAS
aluguel_casa[["area","rooms","bathroom","parking_spaces","rent_amount","property_tax","fire_insurance","total"]].describe()
#COUNT = CONTAGEM DE LINHAS
#MEAN = MÉDIA
#STD = DESVIO PADRÃO
#MIN = MENOR VALOR DA COLUNA
```

```python
#FILTANDRO CASAS COM MALO TOTAL ATÉ $4.000
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000)]
aluguel_casa.describe()
```

```python
#ADICIONANDO O FILTRO DE ANIMAIS
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept')]
aluguel_casa.head()
```

```python
#ADICIONANDO O FILTRO COM 2 OU 3 QUARTOS
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept') & (aluguel_casa['rooms']==2)|(aluguel_casa['rooms']==3)]
aluguel_casa.head()
```

```python
#FAZENDO O MESMO FILTRO ACIMA, MAS COM .QUERY
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3"))
aluguel_casa.head()
```

```python
#AGRUPANDO DADOS POR CIDADE E PEGANDO A MÉDIA DO TOTAL DE CADA UMA
agg_city_price = aluguel_casa.groupby(['city'])['total'].agg('mean').reset_index()
agg_city_price
```

```python
#AGRUPANDO DADOS POR QUANTIDADE DE BANHEIRO E QUARTOS, MOSTRANDO A MÉDIA TOTAL
agg_bath_room_price = aluguel_casa.groupby(['bathroom','rooms']) ['total'].agg('mean').reset_index()
agg_bath_room_price
```

```python
#IMPORTANDO A BIBLIOTECA SEABORN
import seaborn as sns
```

```python
# UTILIZANDO O SEABORN PARA FAZER O GRAFICO DE MÉDIA DE PREÇOS, COM BASE NA QUANTIDADE DE QUARTOS
ax =sns.barplot(data= agg_bath_room_price, x= 'rooms', y= 'total', color = 'blue');
ax.set(xlabel = 'qtd quartos', ylabel = 'media de preco');
```

```python
# UTILIZANDO O SEABORN PARA FAZER O GRAFICO DE MÉDIA DE PREÇOS, COM BASE NA QUANTIDADE DE BANHEIROS
ax =sns.barplot(data= agg_bath_room_price, x= 'bathroom', y= 'total', color = 'green');
ax.set(xlabel = 'qtd banheiros', ylabel = 'media de preco');
```

```python
#CRIANDO A VARIÁVEL AREA
agg_area = aluguel_casa[['area']]
agg_area.describe()
```

```python
#LIMITANDO A ÁREA ATÉ 100(75% DO TOTAL)
agg_area = agg_area.query('area <= 100')
agg_area.describe()
```

```python
#UTILIZANDO SEABORN PARA FAZER O GRÁFICO COM BASE NA ÁREA
sns.histplot(data = agg_area, x = 'area');
```

```python
#UTILIZANDO O SEABORN PARA FAZER O GRÁFICO TOTAL POR CIDADE
sns.histplot(data = aluguel_casa, x = 'total', hue ='city');
```

```python
#FILTANDO APENAS PELA CIDADE DO PORTO
aluguel_casa = aluguel_casa.query("city == 'Porto' ")
aluguel_casa.head()
```

```python
#CRIANDO A VARIÁVEL ANDAR
agg_floor_price = aluguel_casa.groupby("floor")['total'].agg('mean').reset_index()
agg_floor_price
```

```python
#UTILIZANDO O SEABORN PARA FAER O GRÁFICO POR ANDAR
sns.barplot(data = agg_floor_price, x ='floor', y = 'total', color = 'blue');
```
